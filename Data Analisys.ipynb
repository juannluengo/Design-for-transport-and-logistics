```python
import pandas as pd
import matplotlib.pyplot as plt
# Read Excel file into DataFrame
df = pd.read_excel('/Users/fernandomarti/Desktop/A1_1500_1.xlsx')

# Display first few rows of the DataFrame
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Split DataFrame into pickups and deliveries
pickups = df[df['Type'] == 3]  # Type 3 represents pickups
deliveries = df[df['Type'] == 2]  # Type 2 represents deliveries

# Plot pickups and deliveries with smaller points
plt.scatter(pickups['X'], pickups['Y'], color='red', label='Pickups', s=10)  # Adjust size with 's' parameter
plt.scatter(deliveries['X'], deliveries['Y'], color='blue', label='Deliveries', s=10)  # Adjust size with 's' parameter

# Set plot labels and title
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Positions of Pickups and Deliveries')

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python
# VISUALIZE HOW THE DEMAND IS DISTRIBUTED OVER THE DAYS 


# Group orders by Day-order and count the number of orders for each day
order_counts_per_day = df.groupby('Day-Order').size()

# Plot bar chart
plt.bar(order_counts_per_day.index, order_counts_per_day.values)

# Set plot labels and title
plt.xlabel('Day')
plt.ylabel('Number of Orders')
plt.title('Distribution of Orders over Days')

# Show plot
plt.show()
```
